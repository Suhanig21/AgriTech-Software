```python
import numpy as np
%pip install pandas
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("Datasets\\Fertilizer Prediction.csv")
data = data.drop(columns = ["Crop Type","Soil Type"], axis = 1)
data.head()
```

```python
data.columns
```

```python
features = ['N', 'K', 'P', 'temperature', 'moisture','Humidity']

X = data[features]
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
pca = PCA(n_components = len(features))
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_ratio
```

```python
eigenvectors = pca.components_
eigenvectors
```

```python
weights = eigenvectors[0] * explained_variance_ratio[0]
for i in range(1, len(features)):
  weights += eigenvectors[i] * explained_variance_ratio[i]
weights
```

```python

weights /= weights.sum()
weights
```

```python
def calculate_sqi(X_scaled, weights):
  return (X_scaled * weights).sum(axis = 1)
```

```python
sqi= calculate_sqi(X_scaled, weights)
sqi
```

```python
minmax = MinMaxScaler()
sqi_scaled = minmax.fit_transform(sqi.reshape(-1, 1))
sqi_scaled
```

```python
data['SQI'] = sqi_scaled
data.head()
```

```python
print(min(data['SQI']))
print(max(data['SQI']))
```

```python
from imblearn.over_sampling import SMOTE
# from google.colab import files

X = fert_data.drop(columns = ['Fertilizer Name'], axis = 1)
y = fert_data['Fertilizer Name']

smote = SMOTE(random_state = 42)
X_resampled, y_resampled = smote.fit_resample(X, y)

df_resampled = pd.concat([pd.DataFrame(X_resampled, columns = X.columns), pd.Series(y_resampled, name = 'Fertilizer')], axis = 1)
print(df_resampled.head())
print(df_resampled.value_counts('Fertilizer'))
print(df_resampled)
# df_resampled.to_csv('New Crop Dataset.csv', index = False)
# files.download('New Crop Dataset.csv')
```

```python
X = df_resampled.drop("Fertilizer", axis = 1)
Y = df_resampled["Fertilizer"]

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier(n_estimators = 100, random_state = 42)
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(Y_test, y_pred, zero_division = 0))
```

```python
data['Fert_Predict'] = model.predict(fert_data.drop(columns = ["Fertilizer Name"]))
data
```
